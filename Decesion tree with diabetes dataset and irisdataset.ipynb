```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
iris
```

```python
print(iris['DESCR'])
```

```python
print(iris['target'])
```

```python
#independent variable
X=pd.DataFrame(iris['data'],
               columns=['sepal length in cm','sepal width in cm',
                        'petal length in cm','petal width in cm'])
```

```python
#dependent varaible
y=pd.DataFrame(iris['target'],columns=['type'])
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
## apply decision tree  classifier 
from sklearn.tree import DecisionTreeClassifier
DT=DecisionTreeClassifier()
```

```python
DT.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure (figsize=(15,10))
tree.plot_tree(DT,filled=True)
```

```python
y_pred=DT.predict(X_test)
from sklearn.metrics import confusion_matrix,classification_report
cm=confusion_matrix(y_test,y_pred)
cr=classification_report(y_test,y_pred)
print(cm)
print(cr)
```

```python
##hyper parameter tuning
from sklearn.tree import DecisionTreeClassifier
DT=DecisionTreeClassifier( max_depth=2)
DT.fit(X_train,y_train)
from sklearn import tree
plt.figure (figsize=(15,10))
tree.plot_tree(DT,filled=True)
```

```python
y_pred=DT.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
cm=confusion_matrix(y_test,y_pred)
cr=classification_report(y_test,y_pred)
print(cm)
print(cr)
```

### Decdesion tree classifier preprunning and hyperparameter tunning

```python
params={
    'criterion':['gini', 'entropy', 'log_loss'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5,6,7],
    'max_features':['auto','squrt','log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
grid= GridSearchCV(DT,param_grid=params,verbose=3,cv=5,scoring='accuracy')
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred=grid.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
cm=confusion_matrix(y_test,y_pred)
cr=classification_report(y_test,y_pred)
print(cm)
print(cr)
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(y_test,y_pred)
```

### DECESION TREE REGRESSOR 

```python
from sklearn.datasets import load_diabetes
```

```python
dataset=load_diabetes()
```

```python
print(dataset['DESCR'])
```

```python
dataset
```

```python
X=pd.DataFrame(dataset['data'],columns=['age','sex','bmi','bp','s1','s2','s3','s4','s5','s6'])
```

```python
X
```

```python
y=dataset['target']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
X_train.corr()
```

```python
sns.heatmap(X_train.corr())
```

```python
from sklearn.tree import DecisionTreeRegressor
DT=DecisionTreeRegressor()
```

```python
DT.fit(X_train,y_train)
```

```python
y_pred=DT.predict(X_test)
```

```python
params={
    'criterion':['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5,6,7],
    'max_features':['auto','squrt','log2']
}
from sklearn.model_selection import GridSearchCV
grid= GridSearchCV(DT,param_grid=params,verbose=3,cv=5,scoring='neg_mean_squared_error')
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred=grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```

```python
DT=DecisionTreeRegressor(criterion='friedman_mse',max_depth=3,max_features='log2',splitter='best')
```

```python
import warnings
warnings.filterwarnings('ignore')
DT.fit(X_train,y_train)
```

```python
y_pred=DT.predict(X_test)
```

```python
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```

```python
from sklearn import tree
plt.figure (figsize=(15,10))
tree.plot_tree(DT,filled=True)
```
