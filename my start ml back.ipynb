```python
import pandas as pd 
import numpy as np 
from sklearn.datasets import load_boston
import matplotlib.pyplot as plt 
```

```python
df = load_boston()
```

```python
 df
```

```python
dataset = pd.DataFrame(df.data)
```

```python
dataset
```

```python
dataset.columns = df.feature_names
```

```python
dataset
```

```python
df.target
```

```python
dataset['price'] = df.target
```

```python
dataset
```

```python
x = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1:]
```

```python
y
```

```python
from sklearn.model_selection import cross_val_score
from sklearn .linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
mse = cross_val_score  (lr,x,y,scoring='neg_mean_squared_error',cv =5)
```

```python
mean_mse = np.mean(mse)
```

```python
mean_mse
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn .linear_model import Ridge
ridge = Ridge()
parameter = {"alpha":[1e-15,1e-10,1e-3,1e-2,1,5,10,20,30,35,45,40,50,55,100]}
mse=GridSearchCV (ridge,parameter,scoring='neg_mean_squared_error',cv =10)
mse.fit(x,y)
```

```python
print(mse.best_params_)
print(mse.best_score_)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn .linear_model import Lasso
lasso = Lasso()
parameter = {"alpha":[1e-15,1e-10,1e-3,1e-2,1,5,10,20,30,35,45,40,50,55,100]}
mse=GridSearchCV (lasso,parameter,scoring='neg_mean_squared_error',cv =5)
mse.fit(x,y)
```

```python
print(mse.best_params_)
print(mse.best_score_)
```

```python
import statsmodels.api as sm
x = sm.add_constant(x)
```

```python
model = sm.OLS(y,x).fit()
```

```python
import statsmodels.api as sm
x = sm.add_constant(x)
model = sm.OLS(y,x).fit_regularized(method='elastic_net', alpha=0.1, L1_wt=0)
regularized_regression_parameters = model.params
print(regularized_regression_parameters)
```
