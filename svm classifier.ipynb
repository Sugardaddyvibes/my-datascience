## Hyperparameter tunning with svc

```python
from sklearn.model_selection import GridSearchCV

param_grid={ 'C':[0.1,1,100.1000],
            'kernel':['rbf','poly','sigmoid'],
            'gamma':[1,0.1,0.001,0.01,0.0001]}
```

```python
grid=GridSearchCV(SVC(),param_grid=param_grid,verbose=3,refit=True)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_ 
```

```python
y_pred=grid.predict(X_test)
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
X=df[['tip', 'sex', 'smoker', 'day', 'time', 'size']]
y=df['total_bill']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.4)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le1=LabelEncoder()
le2=LabelEncoder()
le3=LabelEncoder()
```

```python
le1.fit_transform(X_train['sex'])
le2.fit_transform(X_train['smoker'])
le3.fit_transform(X_train['time'])
```

```python
X_train['sex']=le1.fit_transform(X_train['sex'])
X_train['smoker']=le2.fit_transform(X_train['smoker'])
X_train['time']=le3.fit_transform(X_train['time'])
```

```python
X_train
```

```python
X_test['sex']=le1.fit_transform(X_test['sex'])
X_test['smoker']=le2.fit_transform(X_test['smoker'])
X_test['time']=le3.fit_transform(X_test['time'])
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct=ColumnTransformer(transformers=[('onehot', OneHotEncoder(drop='first'),3)],remainder='passthrough')
```

```python
ct=ColumnTransformer(transformers=[('onehot', OneHotEncoder(drop='first'),[3])],remainder='passthrough')
import sys
np.set_printoptions(threshold=sys.maxsize)
X_train=ct.fit_transform(X_train)
```

```python
X_test=ct.fit_transform(X_test)
```

```python
X_test
```

```python
from sklearn.svm import SVR
svr =SVR()
svr.fit(X_train,y_train)
```

```python
y_pred=svr.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```
